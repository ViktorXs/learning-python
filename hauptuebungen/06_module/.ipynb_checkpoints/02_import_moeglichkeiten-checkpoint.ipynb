# Standard: <code>import</code> <code>as</code>

Die Funktion <code>pyplot</code> von <code>matplotlib</code> laden:

```python
import matplotlib.pyplot as plt  # Funktion in plt verkürzt / umbenannt.

plt.plot([1, 2, 3], [2, 1, 5])  # Das Schreiben der vollständigen Funktion matplotlib.pyplot.plot nicht nötig.
pyplot.show()
```

# <code>import</code>

Die Funktion <code>matplotlib.pyplot</code> direkt laden:

```python
import matplotlib.pyplot

matplotlib.pyplot.plot([1, 2, 3], [1, 2, 4])
pyplot.show()
```

# <code>from</code> <code>import</code>

```python
from matplotlib import pyplot
# from matplotlib import pyplot as plt  # Zur Verhinderung von Überschreibungen, das Modul umbenennen.

pyplot.plot([1, 2, 3], [5, 4, 6])
pyplot.show()
```
