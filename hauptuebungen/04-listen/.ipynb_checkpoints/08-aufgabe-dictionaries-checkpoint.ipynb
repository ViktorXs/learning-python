# Aufgabe: Dictionaries

Die Datei "/names.csv" einlesen, berechnen welcher Name insgesamt in den gesamten USA am Häufigsten vergeben wurde.

- Zuerst Daten in Dictionary einlesen, zählen, wie oft jeder Vorname insgesamt vorkommt.
- Dictionary analysieren und den häufigsten Namen finden. 
- Gegebenenfalls Strings in Zahlen umwandeln.
- Gesamten Code in einer Zelle halten.

## Code

```python
def namesOccurrence():
    with open("../data/names.csv", "r") as file:

        names = {}

        for line in file:
            splitted = line.strip().split(",")
            if splitted[0] == "Id":
                continue

            name = splitted[1]
            freq = int(splitted[5])

            if name in names:
                names[name] = names[name] + freq     # Wenn der Name öfters auftaucht, zähle ihn mit dem Wert zusammen
            else:
                names[name] = freq     # Ansonsten setze den Schlüssel auf den Wert (freq)

        maxOccurrences = 0
        name = ""

        for key, value in names.items():   # Die Daten in der dictionary mit key und value durchgehen .items()
            if maxOccurrences < value:     # Wenn maxOccurrences (der Name) kleiner als der Wert, der gerade betrachtet wird, dann...
                maxOccurrences = value     # ... aktualisiere das Höchste, was gefunden wurde und ...
                name = key                 # ... setze die Variable auf den häufigsten gefundenen Namen
                count = value
            print([key, value])
        print("The most given name is " + name + " and is used " + str(count) + " times.")
```

```python
namesOccurrence()
# Toll, um interessante Namen zu finden. Z.B. suche nach dem zweiten Eintrag mit "Lord" oder Jamesryan usw...
```
