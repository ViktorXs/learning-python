# <code>finally</code>
Nach einem Fehler sauber aufräumen.
Erzwingen durch <code>finally</code>.

Was definitiv noch ausgeführt werden muss, weil z.B. files, die mit <code>file = open</code> geöffnet werden, sich nach einem fehler nicht schließen und offen bleiben.

```python
try:
    file = open("existiert_nicht.txt", "r")
    print(file)
    print(5 / 0)

except FileNotFoundError:
    print("Datei wurde nicht gefunden")

finally:  # Egal ob Fehler auftritt oder nicht, der Code hiernach wird definitv ausgeführt.
    print("finally")
    file.close()
```

## Exceptions in Python

In dieser Lektion lernst du, wie du:

- nach einem Fehler sauber "aufräumen" kannst,
- dies mit einem "finally" korrekt erzwingen kannst.

```python
try:
    file = open("existiert.txt", "r")
    print(file)
    print(5 / 0)
except FileNotFoundError:
    print("Datei wurde nicht gefunden")
finally:
    print("FINALLY!!!")
    file.close()
```
