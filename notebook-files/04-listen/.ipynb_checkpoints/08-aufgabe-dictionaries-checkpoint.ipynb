# Aufgabe: Dictionaries

Die Datei "/names.csv" einlesen, berechnen welcher Name insgesamt in den gesamten USA am Häufigsten vergeben wurde.

- Zuerst Daten in Dictionary einlesen, zählen, wie oft jeder Vorname insgesamt vorkommt.
- Dictionary analysieren und den häufigsten Namen finden. 
- Gegebenenfalls Strings in Zahlen umwandeln.
- Gesamten Code in einer Zelle halten.

## Code

```python
def listData():    
    with open("../data/names.csv", "r") as file:
        
        names = {}
        count = 0
        
        for line in file:
            splitted = line.strip().split(",")
            if splitted[0] == "Id":
                continue
            
            name = splitted[1]
            freq = int(splitted[5])
            
            if name in names:
                names[name] = names[name] + count
            else:
                names[name] = count
listData()
```

```python
print(names)
```

```python
for key, value in names.items():
    print([key, value])
```

## Schmierblatt

```python
with open("../data/names.csv", "r") as file:
    for line in file:
        clean = line.strip().split(",")
        if clean[0] == "Id":
            continue
        print(clean)
        break
```

```python
liste = ["A", "E", "A", "C", "E", "A", "A", "E", "F", "E", "A"]

d = {}

for element in liste:
    if element in d:
        d[element] = d[element] + 1
    else:
        d[element] = 1
print(d)
```

```python
occurrence = 0

for key, value in d.items():
    if occurrence < value:
        occurrence = value
print(occurrence)
```
